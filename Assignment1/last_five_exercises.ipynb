```python
# Question 6:
# With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.
import pandas as pd

data={'category':['a','b','c','d','d','b','b','c','d','a'],'column1':[1,5,3,12,78,56,14,67,89,90],'column2':[45,67,23,75,27,4,8,97,89,45]}
df=pd.DataFrame(data)
print(df)

sum=df.groupby('category')['column1'].sum()
print(sum)
sum2=df.groupby('category')['column2'].sum()
print(sum2)
```

```python
# Question 9:
# With Pandas, perform a merge operation between two DataFrames based on a common column.
first_data={'Flowers':['rose','lily','lotus'],'color':['pink','yellow','blue']}
df1=pd.DataFrame(first_data)

second_data={'color':['red','yellow','purple'],'Fruit':['orange','apple','mango']}
df2=pd.DataFrame(second_data)

merge=pd.merge(df1,df2, on='color')
print(merge)
```
