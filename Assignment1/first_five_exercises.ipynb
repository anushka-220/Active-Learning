```python
# Question 2:
# With Pandas, read a CSV file into a DataFrame. Then, filter the rows where a specific column's value is greater than a certain threshold.
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()
ir_df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
ir_df.to_csv('ir_dataset.csv',index= False)
print(ir_df)
load_ir_df=pd.read_csv('ir_dataset.csv')

# let the specific column be petal length and threshold be 5.5
after_filter=load_ir_df[load_ir_df['petal length (cm)']> 5.5]
print(after_filter)
```

```python
# Question 3:
# Using Matplotlib, plot a line graph showing the trend of a numerical dataset.
# Customize the plot with title, axis labels, and a legend.
import matplotlib.pyplot as plt

x= ir_df['sepal length (cm)']
y=ir_df['sepal width (cm)']

plt.plot(x, label='Sepal Length', color='red')
plt.plot(y, label='Sepal Width', color='blue', linestyle='-')

plt.title('Graph of Sepal Width and Sepal Length')
plt.xlabel('Values')
plt.ylabel('Length or Width(in cm)')
plt.legend()
plt.show()
```

```python
# Question 4:
# Create a Pandas DataFrame with random values and perform the following operations:
# - Replace NaN values with the mean of the column.
# - Sort the DataFrame based on two columns.

import pandas as pd
import numpy as np

data={'series':[1,5,6,3,9,10,np.nan,46,49,np.nan,np.nan],'random values':[12,56,43,np.nan,5,9,np.nan,34,76,np.nan,2],}
df=pd.DataFrame(data)
print(df)

df.fillna(df.mean(), inplace=True)
print(df)
df.sort_values(by=['series','random values'])
print(df)
```

```python
# Question 5:
# Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

# I have chosen a square matrix for determinant calculation
matrix_1=np.array([[2,3], [7,8]])
matrix_2=np.array([[5,6],[6,2]])
dot=np.dot(matrix_1,matrix_2)
print("Dot product:",dot)
det=np.linalg.det(dot)
print("Determinant: ",det)
```
